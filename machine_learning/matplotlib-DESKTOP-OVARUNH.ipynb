```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
from scipy import datasets
```

```python
x = np.linspace(0, 10, 15)
y = x**2
```

Il faut que les deux variables soit correspondantes (de mÃªme taille).

```python
plt.plot(x, y, c='red', lw=3, ls='--') # lw (line width) ls (line style)
# plt.scatter(x, y)
```

### Les basiques

```python
plt.figure(figsize=(6, 4)) # DÃ©but de la figure : la fenÃ©tre de travail sur laquel travailler. Attribue figsize pour rÃ©gler la taille.
plt.plot(x, y, label='quadratique') # Contenu 1
plt.plot(x, x**3, c='red', label='cubique') # Contenu 2â¦
plt.xlabel('X') # Les labels
plt.ylabel('Y')
plt.title("X vs X**2")
plt.legend() # Pour utiliser cet argument, il faut le prÃ©ciser dans la crÃ©ation du graphique
plt.show() # Affiche la figure
```

### Les subplot's

```python
f = lambda x, y: np.sin(x) + np.cos(x + y)*np.cos(x)
```

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

## Contours

```python
# Visualisation du dessus, utile pour observer la minimasition du model, observer les minimums locaux et globaux
plt.contour(X, Y, Z, 20, colors='black') # Le niveau de profondeur et colors black, trait plein positif, tiret nÃ©gatif
# plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```
